```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import datetime
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
df = pd.read_csv('final_data.csv')
```

```python
print(df['price'].value_counts())
```

```python
price_regex = r'^\$?([\d,]+)(?:\.(\d{2}))?$'

# Define function to extract numeric value from price string
def extract_numeric_value(price_str):
    match = re.match(price_regex, price_str)
    if match:
        return float(match.group(1).replace(',', '') + '.' + (match.group(2) if match.group(2) else '00'))
    else:
        return None
```

```python
df['price'] = df['price'].astype(str)
```

```python
df['price'] = df['price'].apply(extract_numeric_value)
```

```python
print(df['price'].value_counts())
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

# Remove outliers from the DataFrame
df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
```

```python
plt.boxplot(df['price'])
```

```python
df.shape
```

```python
re.search('\d+', '< 3 minutes ago')
```

```python
def parse_date_string(date_str):
    if "minutes ago" in date_str or "hours" in date_str:
        ptrn = re.search('\d+', date_str)
        minutes_ago = int(ptrn.group(0))
        return datetime.datetime.now() - datetime.timedelta(minutes=minutes_ago)
    elif "yesterday" in date_str.lower():
        return datetime.datetime.now() - datetime.timedelta(days=1)
    elif "today" in date_str.lower():
        return datetime.datetime.now()
    else:
        return pd.to_datetime(date_str)

df['date_posted'] = df['date_posted'].apply(parse_date_string)
df['month'] = df['date_posted'].dt.month
```

```python
sns.relplot(y='date_posted', x='price', hue='month', kind='scatter', data=df)
```

```python
# Extract city name from address column
def extract_city(address):
    cities = ['brampton', 'mississauga', 'don mills', 'toronto', 'ajax', 'markham', 'scarborough', 'north york']
    for city in cities:
        if city in address.lower():
            return city
    return 'Other'

df['city'] = df['address'].apply(extract_city)

grouped = df.groupby('city')
for name, group in grouped:
    print(f"{name}: {len(group)}")
```

```python
ax = sns.boxplot(x='city', y='price', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
```

```python
sns.relplot(y='date_posted', x='price', hue='city', kind='scatter', data=df, height=10)
```

```python
# pattern detecting male and female
pattern = re.compile(r'\b(male|female|m|f|man|woman)\b', flags=re.IGNORECASE)

# Apply the pattern
df['gender'] = df['description'].str.extract(pattern, expand=False)

gender_map = {
    'male': 'male',
    'Male': 'male',
    'MALE': 'male',
    'm': 'male',
    'man': 'male',
    'female': 'female',
    'Female': 'female',
    'FEMALE': 'female',
    'f': 'female',
    'woman': 'female'
}
# Group using gender
df['gender'] = df['gender'].map(gender_map)
counts = df['gender'].value_counts()
```

```python
plt.pie(counts.values, labels=counts.index, autopct="%0.2f%%", startangle=90, shadow=True)
plt.title('Number of Rooms defined for gender specific')
plt.show()
```

```python
# Filter the DataFrame to include only listings that contain 'basement' (case-insensitive)
basement_df = df[df['description'].str.contains('basement', case=False) == True]

# Filter the DataFrame to include only listings that contain 'main floor' (case-insensitive)
main_floor_df = df[df['description'].str.contains('main floor', case=False) == True]

# Count the number of listings that contain 'basement' and 'main floor'
basement_count = basement_df.shape[0]
main_floor_count = main_floor_df.shape[0]

# Create a bar plot of the counts
counts = [basement_count, main_floor_count]
labels = ['Basement', 'Main Floor']

plt.figure(figsize=(10, 12))
plt.bar(labels, counts, width=0.3)
plt.title('Number of Listings with Basement and Main Floor')
plt.xlabel('Type of Unit')
plt.ylabel('Number of Listings')
plt.show()
```

```python
df['num_beds'] = df['title'].str.extract('(\d+)\s*Bed', flags=re.IGNORECASE, expand=False).fillna(-1).astype(int)
```

```python
num_beds = df['num_beds'].value_counts()
num_beds = num_beds[(num_beds.index < 7) & (num_beds.index > 0)]

plt.bar(num_beds.index.astype(str), num_beds.values)
plt.title('Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.show()
```
