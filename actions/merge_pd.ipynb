```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
df = pd.read_csv('rental_canada.csv')
df.head()
```

```python
# https://www.kijiji.ca append to all url value
df['url'] = 'https://www.kijiji.ca' + df['url']
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.to_csv('rental_canada_new.csv')
```

```python
df1 = pd.read_csv('kijiji_toronto_apartments_300.csv')
df2 = pd.read_csv('kijiji_toronto_apartments_new.csv')
df3 = pd.concat([df1, df2])
```

```python
df3['description'] = df3['description'].str.replace('Description', '')
df3.head()
```

```python
# convert date_posted to dd/mm/yyyy format
df3['date_posted'] = pd.to_datetime(df3['date_posted'])
df3['date_posted'] = df3['date_posted'].dt.strftime('%d/%m/%Y')
df3.head()
```

```python
# Extract city name from address column
def extract_city(address):
    cities = ['brampton', 'mississauga', 'don mills', 'toronto', 'ajax', 'markham', 'scarborough', 'north york']
    for city in cities:
        if isinstance(address, str):
            if city in address.lower():
                return city.title()
    return 'Other'

df3['address'] = df3['address'].apply(extract_city)

grouped = df3.groupby('address')
for name, group in grouped:
    print(f"{name}: {len(group)}")
```
